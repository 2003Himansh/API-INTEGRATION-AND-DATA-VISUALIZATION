```python
import requests
import matplotlib.pyplot as plt
import seaborn as sns

API_KEY = '182a2c4afaeb8957bba910bc37567642'  
city_name = 'faridabad'
URL = f'https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_KEY}&units=metric'


response = requests.get(URL)

if response.status_code == 200:
    data = response.json()
    
    # Extract weather details
    temp = data['main']['temp']
    feels_like = data['main']['feels_like']
    min_temp = data['main']['temp_min']
    max_temp = data['main']['temp_max']
    weather_desc = data['weather'][0]['description']

    print(f"Current Weather in {city_name}: {weather_desc}")
    print(f"Temperature: {temp}째C")
    print(f"Feels Like: {feels_like}째C")
    print(f"Min Temperature: {min_temp}째C")
    print(f"Max Temperature: {max_temp}째C")

    # Step 3: Data Visualization
    temperatures = [min_temp, temp, max_temp, feels_like]
    labels = ['Min Temp', 'Current Temp', 'Max Temp', 'Feels Like']

    #sns.set_style("whitegrid")
    colors=sns.color_palette("coolwarm",len(labels))
    plt.figure(figsize=(8, 5))
    sns.barplot(x=labels, y=temperatures, hue=labels, palette=colors, legend=False)


    plt.ylabel('Temperature (째C)')
    plt.title(f'Temperature Variations in {city_name}')
    plt.show()

else:
    print(f"Error fetching data. Status code: {response.status_code}")
```
